```python
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos#Top_videos")
```

```python
Rank_R = []
video_name = []
Artist_A= []
Upload_date=[]
Views_V=[]
```

```python
#scraping video_name from the given page             #finde the Lcators:
video_tags = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[2]')
for i in video_tags:
    video =i.text
    video_name.append(video)
    
#scraping from Rank the given page
Rank_list = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[1]')
for i in Rank_list:
    Rank = i.text
    Rank_R.append(Rank)
    
#scraping from Artist from the given page
Artist_tags = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[3]')
for i in Artist_tags:
    Artist = i.text
    Artist_A.append(Artist)
    
#scraping upload_date from the given page
Upload_list = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[5]')
for i in Upload_list:
    Upload = i.text
    Upload_date.append(Upload)
    
    
#scraping upload_date from the given page
Views_list = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[4]')
for i in Views_list:
    Views = i.text
    Views_V.append(Views)
    
```

### make a Data Frame:

```python
import pandas as pd
df = pd.DataFrame({'Rank':Rank_R ,'Video_Name':video_name,'Artist':Artist_A, 'Views':Views_V,'Upload_date':Upload_date})
df
```

```python
search = driver.find_element(By.XPATH,'//input[@type="text"]')
search_button = driver.find_element(By.XPATH,'//input[@type="submit"]')
```

```python
#make a list:
Dataset_name =[]
Data_type = []
Task_T=[]
Attribute_type =[]
No_instances =[]
No_attribute =[]
Year_R=[]
```

```python
#scraping the dataset name from gven page
Dataset_tag = driver.find_elements(By.XPATH,'//table[@border="1"]/tbody/tr/td[1]')
for i in Dataset_tag:
    Dataset =i.text
    Dataset_name.append(Dataset)
#scraping the data type from given page
Data_tag = driver.find_elements(By.XPATH,'//table[@border="1"]/tbody/tr/td[2]')
for i in Data_tag:
    Data =i.text
    Data_type.append(Data)
#scraping task frome given page
task_tag = driver.find_elements(By.XPATH,'//table[@border="1"]/tbody/tr/td[3]')
for i in task_tag:
    task =i.text
    Task_T.append(task)
#scraping Attribute type from given page
Attribute_tag = driver.find_elements(By.XPATH,'//table[@border="1"]/tbody/tr/td[4]')
for i in Attribute_tag[0:623]:
    Attribute=i.text
    Attribute_type.append(Attribute)
#scraping no. of instance from given page
instances_tag = driver.find_elements(By.XPATH,'//table[@border="1"]/tbody/tr/td[5]')
for i in instances_tag:
    instances=i.text
    No_instances.append(instances)
#scraping no. of attributes fom given page
Attribute_list = driver.find_elements(By.XPATH,'//table[@border="1"]/tbody/tr/td[6]')
for i in Attribute_list:
    Attribute=i.text
    No_attribute.append(Attribute)
#scraping year rom the given page
Year_tag = driver.find_elements(By.XPATH,'//table[@border="1"]/tbody/tr/td[7]')
for i in Year_tag:
    Year=i.text
    Year_R.append(Year)

```

### make a dataframe:

```python
import pandas as pd
df = pd.DataFrame({'Dataset_name':Dataset_name,' Data_type': Data_type,'Task_T':Task_T,'Attribute_type':Attribute_type,'No_instances':No_instances,' No_attribute':No_attribute,'Year_R': Year_R})
df
```
